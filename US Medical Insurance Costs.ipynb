# U.S. Medical Insurance Costs

The goals in this project will be to analyze various atributes in a CSV file of a medical insurance costs, to learn more about the patient information in the file using Python fundamentals. 

```python
# Import csv library
import csv
```

For this project we will import only csv library in order to work with insurance.csv. We could work with other libraries, but for this project csv library must be enough.

In order to plan how to import the data, we need to look of some information about the data. The name of the columns, the lenght of the data, if there are any missing values, the type of the values (categorical of numerical).

```python
# Create a function to import de csv file

def import_csv():
    with open('insurance.csv') as i_file:
        csv_file = csv.DictReader(i_file)
        data = [f for f in csv_file]
        return data
    
global average_age
global average_charge
data_d = import_csv()
```

```python
# Creating a class Patient Info
class PatientInfo:
    def __init__(self):
        self.patient_dict = {}
        self.northeast_females_smokers = 0
        self.northwest_females_smokers = 0
        self.southeast_females_smokers = 0
        self.southwest_females_smokers = 0
        self.northeast_males_smokers = 0
        self.northwest_males_smokers = 0
        self.southeast_males_smokers = 0
        self.southwest_males_smokers = 0
        self.females_smokers = 0
        self.males_smokers = 0
        self.charges = []
        self.total_charges = 0
        self.regions_unique = []
        self.females = 0
        self.males = 0
        self.tot_ages = 0
        self.age_list = []
    
    def average_age(self):
        global average_age
        for age in data_d:
            self.age_list.append(age['age'])
            self.tot_ages += int(age['age'])
            average_age = float(round(self.tot_ages / len(self.age_list), 2))
        return f"Average Age: {average_age}"
    
    def count_sex(self):
        for sex in data_d:
            if sex['sex'][0] == 'm':
                self.males += 1
            else:
                self.females += 1
        return f"Males: {self.males}\nFemales: {self.females}"
    
    def unique_regions(self):
        for region in data_d:
            if region['region'] not in self.regions_unique:
                self.regions_unique.append(region['region'])
        return f"Unique regions:\n{self.regions_unique}"
    
    def average_charges(self):
        global average_charge
        for charge in data_d:
            self.total_charges += float(charge['charges'])
            self.charges.append(charge['charges'])
            average_charge = round(self.total_charges / len(self.charges))
        return f"Average charges: {average_charge}"
    
    def analyze_smokers(self):
        for smokers in data_d:
            if smokers['sex'] == 'male' and smokers['smoker'] == 'yes':
                self.males_smokers += 1
            if smokers['sex'] == 'female' and smokers['smoker'] == 'yes':
                self.females_smokers += 1
        return f"Total males smokers: {self.males_smokers}\nTotal females smokers: {self.females_smokers}"
    
    def analyze_smoker_by_region(self):
        for s in data_d:
            if s['sex'] == 'male' and s['smoker'] == 'yes' and s['region'] == 'southwest':
                self.southwest_males_smokers += 1
            elif s['sex'] == 'female' and s['smoker'] == 'yes' and s['region'] == 'southwest':
                self.southwest_females_smokers += 1
            elif s['sex'] == 'male' and s['smoker'] == 'yes' and s['region'] == 'southeast':
                self.southeast_males_smokers += 1
            elif s['sex'] == 'female' and s['smoker'] == 'yes' and s['region'] == 'southeast':
                self.southeast_females_smokers += 1
            elif s['sex'] == 'male' and s['smoker'] == 'yes' and s['region'] == 'northwest':
                self.northwest_males_smokers += 1
            elif s['sex'] == 'female' and s['smoker'] == 'yes' and s['region'] == 'northwest':
                self.northwest_females_smokers += 1
            elif s['sex'] == 'male' and s['smoker'] == 'yes' and s['region'] == 'northeast':
                self.northeast_males_smokers += 1
            elif s['sex'] == 'female' and s['smoker'] == 'yes' and s['region'] == 'northeast':
                self.northeast_females_smokers += 1
            
        return f"Total males smokers in southwest: {self.southwest_males_smokers}\n" \
               f"Total females smokers in southwest: {self.southwest_females_smokers}\n" \
               f"Total males smokers in southeast: {self.southeast_males_smokers}\n" \
               f"Total females smokers in southeast: {self.southeast_females_smokers}\n" \
               f"Total males smokers in northwest: {self.northwest_males_smokers}\n" \
               f"Total females smokers in northwest: {self.northwest_females_smokers}\n" \
               f"Total males smokers in northeast: {self.northeast_males_smokers}\n" \
               f"Total females smokers in northeast: {self.northeast_females_smokers}" 
    
    def create_dictonary(self):
        for dic in data_d:
            self.patient_dict = {'Age': dic['age'], 'Sex': dic['sex'], 'BMI': dic['bmi'], 
                                 'Number of children': dic['children'], 'Smoker Statuses': dic['smoker'], 
                                 'Region': dic['region'], 'Charges': dic['charges']}
            print(self.patient_dict)
```

```python
patient = PatientInfo()

print(patient.average_age())
```

```python
print(patient.average_charges())
```

```python
print(patient.analyze_smokers())
```

```python
print(patient.analyze_smoker_by_region())
```
